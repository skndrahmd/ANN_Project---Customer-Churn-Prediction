```python
import pandas as pd
import pickle 
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
data = pd.read_csv('Churn_Modelling.csv')
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
df
```

```python
gender_encoder = LabelEncoder()
df['Gender'] = gender_encoder.fit_transform(df['Gender'])
df
```

```python
geo_encoder = OneHotEncoder()
geo_encoded = geo_encoder.fit_transform(df[['Geography']])
geo_encoded = geo_encoded.toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=geo_encoder.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
df = pd.concat([df, geo_encoded_df], axis=1)
df
```

```python
df
```

```python
X = df.drop(['EstimatedSalary'], axis=1)
Y = df['EstimatedSalary']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#ANN train for regression 
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#Build the model

model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)), # HL1 connected to the input layer
    Dense(32, activation='relu'), #HL2
    Dense(1) # Output layer
])

model.compile(optimizer='adam', loss='mean_absolute_error', metrics=['mae'])
```

```python
model.summary()
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
#Train the model on the data
model.fit(X_train, Y_train, epochs=100, batch_size=32, validation_data=(X_test, Y_test), callbacks=[early_stopping])
```

```python
model.save('regression_model.h5')
```
