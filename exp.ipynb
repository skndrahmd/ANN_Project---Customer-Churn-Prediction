```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
#Preprocess the data 
df = df.drop(["RowNumber","CustomerId", "Surname"], axis=1)
df
```

```python
#Encode Categorical variable
label_encoder_gender=LabelEncoder()
df['Gender']=label_encoder_gender.fit_transform(df['Gender'])
df
```

```python
#One hot encode Geography column
from sklearn.preprocessing import OneHotEncoder
geography_encoder = OneHotEncoder()
geo_encoded=geography_encoder.fit_transform(df[["Geography"]])

geo_encoded = geo_encoded.toarray()
geo_columns = geography_encoder.get_feature_names_out(['Geography'])
```

```python
pd.DataFrame(geo_encoded, columns=geo_columns)
```

```python
#Combine all columns

df = pd.concat([df.drop(['Geography'], axis=1), pd.DataFrame(geo_encoded, columns=geo_columns)], axis=1)
df
```

```python
#Save the enoders and scalers in pickle files

with open('label_encoder_gender.pkl', 'wb') as file: 
    pickle.dump(label_encoder_gender, file)

with open('one_hote_encoder_pickle.pkl', 'wb') as file: 
    pickle.dump(geography_encoder, file)
```

```python
#Divide our dataset into independent and dependent features
X = df.drop('Exited', axis = 1)
Y = df['Exited']

#Split data into training and testing sets 

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

#Scale the features 

scaler = StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Save the scaler in pickle file

with open('scaler.pkl', 'wb') as file: 
    pickle.dump(scaler, file)
```

ANN Training

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
```

```python
X_train.shape[1]
```

```python
#Build ANN model

model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)), #HL1 connected to the input layer
    Dense(32, activation="relu"), #HL2
    Dense(1, activation="sigmoid") #Output layer
])
```

```python
model.summary()
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
#Compile the model

model.compile(optimizer=optimizer, loss="binary_crossentropy", metrics=['accuracy'])
```

```python
#Set up tensorboard
from datetime import datetime
log_dir='logs/fit' + datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
#Set up Early stopping
early_stopping_callback=EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)
```

```python
#Train the model
history = model.fit(X_train, Y_train, validation_data=(X_train, Y_train), epochs=100,
                    callbacks=[tensorflow_callback, early_stopping_callback])
```

```python
model.save('model.h5')
```
